```python
# what is wrong?

from shapely.validation import explain_validity, make_valid

explain_validity(india_municipalities[~india_municipalities.is_valid].geometry)
```

```python
# varieties?
india_municipalities['validity']=[x.split('[')[0] for x in india_municipalities.geometry.apply(lambda x: explain_validity(x))]
india_municipalities['validity'].value_counts()
```

```python
india_municipalities.drop(columns=['validity'],inplace=True)

india_municipalities_valid=india_municipalities.copy()

india_municipalities_valid['geometry'] = [make_valid(row)  if not row.is_valid else row for row in india_municipalities_valid['geometry'] ]
#any invalid?
india_municipalities_valid[~india_municipalities_valid.is_valid]
```
